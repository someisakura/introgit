### 類義語検索関数を作成する

```python
def search_similar_words(word):
    #入力した単語がWordnetに存在するかを確認する
    cur = conn.execute("select wordid from word where lemma = '%s'" % word)
    word_id = 99999999
    for row in cur:
        word_id = row[0]
    
    if word_id == 99999999:
        print("Wordnetに存在しない単語です。")
        return
    else:
        print("【「%s」の類義語を出力します。】" % word)
    
    #入力した単語を含む概念を検索
    cur = conn.execute("select synset from sense where wordid = '%s'" % word_id)
    synsets = []
    for row in cur:
        synsets.append(row[0])
    

    #概念に含まれる単語を検索して画面出力する
    num = 1
    for synset in synsets:
        cur1 = conn.execute("select name from synset where synset = '%s'" % synset)
        for row1 in cur1:
            print("%s つめの概念 : %s" % (num, row1[0]))
        
        cur2 = conn.execute("select def from synset_def where (synset = '%s' and lang = 'jpn')" % synset)
        sub_num = 1
        for row2 in cur2:
            print("意味%s : %s" % (sub_num, row2[0]))
            sub_num += 1
        
        cur3 = conn.execute("select wordid from sense where (synset = '%s' and wordid!=%s)" % (synset, word_id))
        sub_num = 1
        for row3 in cur3:
            target_word_id = row3[0]
            _cur3 = conn.execute("select lemma from word where wordid = %s" % target_word_id)
            for _row3 in _cur3:
                print("類義語%s : %s" % (sub_num, _row3[0]))
                sub_num += 1
            
        print("\n")
        num += 1
```

実験

```python
search_similar_words('ネコ')
```

```python
search_similar_words('たんぱく質')
```
